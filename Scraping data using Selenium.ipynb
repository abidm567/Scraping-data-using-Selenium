For this assignment, I'm scraping data from Makemytrip.com, a Hotel booking website. With the help of the scraped data, we can compare the prices and ratings of Hotels for the given dates and take Hotel booking decisions better.

I have selected City as "Hyderabad" and Checkin and Checkout dates as 09-10-2022,10-10-2022 to get the results of Hotels in HYD

Makemytrip has pages where we can scroll without clicking next pages until the results end.

```python
#I'm using range 5 times to scroll the website since it takes more time to load all the results 
#After inspecting the website the elements such as Hotel names, prices, ratings are given as IDs.
#selecting sleep time as 5 and using script to scroll to the bottom of page
```

```python
driver = webdriver.Chrome()
url="https://www.makemytrip.com/hotels/hotel-listing/?checkin=10092022&city=CTHYDERA&checkout=10102022&roomStayQualifier=2e0e&locusId=CTHYDERA&country=IN&locusType=city&searchText=Hyderabad,%20Telangana,%20India&visitorId=d304289f-6e29-4d1b-b1ea-bbe4b41366c5&regionNearByExp=3"
driver.get(url)

for i in range(0,5):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)
hotel_name = driver.find_elements(By.ID,'hlistpg_hotel_name')
for t in hotel_name:
    print(t.text)
```
